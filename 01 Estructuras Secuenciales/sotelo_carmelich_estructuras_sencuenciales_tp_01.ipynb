### Ejercicio 4

4. Crear un programa que pida el radio de un c√≠rculo e imprima su √°rea y su per√≠metro.
f-string (f"texto {valor:.2f}")
:2f convierte el valor flotante a dos decimales en python.

```python
import math
radio = float(input("Ingrese el radio del c√≠rculo: "))
area = math.pi * (radio ** 2)
perimetro = 2 * math.pi * radio
print(f"√Årea: {area:.2f}, Per√≠metro: {perimetro:.2f}")
```

### Ejercicio 7

7. Crear un programa que pida dos n√∫meros enteros distintos de 0 y muestre la suma, divisi√≥n, multiplicaci√≥n y resta.

```python
n1 = int(input("Ingrese el primer n√∫mero (distinto de 0): "))
n2 = int(input("Ingrese el segundo n√∫mero (distinto de 0): "))
print(f"Suma: {n1 + n2}, Resta: {n1 - n2}, Multiplicaci√≥n: {n1 * n2}, Divisi√≥n: {n1 / n2:.2f}")
```

### Ejercicio 8

8. Crear un programa que pida la altura y el peso e imprima el √≠ndice de masa corporal (IMC).

```python
peso = float(input("Ingrese su peso en kg: "))
altura = float(input("Ingrese su altura en metros: "))
imc = peso / (altura ** 2)
print(f"Su IMC es {imc:.2f}")
```

### Ejercicio 9

9. Crear un programa que convierta una temperatura en grados Celsius a Fahrenheit.


```python
celsius = float(input("Ingrese la temperatura en grados Celsius: "))
fahrenheit = (9/5) * celsius + 32
print(f"Equivalente en Fahrenheit: {fahrenheit:.2f}")
```

### Ejercicio 10

10. Crear un programa que pida 3 n√∫meros e imprima su promedio.

```python
num1 = float(input("Ingrese el primer n√∫mero: "))
num2 = float(input("Ingrese el segundo n√∫mero: "))
num3 = float(input("Ingrese el tercer n√∫mero: "))
promedio = (num1 + num2 + num3) / 3
print(f"El promedio es {promedio:.2f}")
```
