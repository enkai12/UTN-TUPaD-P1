### Actividad 6
Utilizando el paquete `statistics`, escribir un programa que tome una lista de n√∫meros aleatorios, calcule moda, mediana y media y determine si hay sesgo positivo, negativo o no hay sesgo.

```python
import random
numeros_aleatorios = [random.randint(1, 100) for i in range(50)]
```

Documentaci√≥n oficial: https://docs.python.org/es/3.8/library/statistics.html

## Definiciones

### Moda

La **moda** es el n√∫mero que aparece con mayor frecuencia en un conjunto de datos.

Ejemplo:
```python
[2, 4, 4, 5] ‚Üí Moda: 4
```

---

### Mediana

La **mediana** es el n√∫mero que se encuentra en el centro cuando los datos est√°n ordenados.

- Si hay cantidad impar de elementos, es el valor central.
- Si hay cantidad par, es el promedio de los dos del medio.

Ejemplos:
```python
[1, 3, 5] ‚Üí Mediana: 3  
[1, 3, 5, 7] ‚Üí Mediana: (3+5)/2 = 4
```

---

### Media

La **media** (promedio) se calcula sumando todos los valores y dividiendo por la cantidad total.

Ejemplo:
```python
[1, 2, 3] ‚Üí Media: (1+2+3)/3 = 2
```

---

### Sesgo

El **sesgo** (o asimetr√≠a) indica c√≥mo est√°n distribuidos los datos:

- **Sesgo positivo**: muchos valores bajos y algunos muy altos.  
  Resultado: `media > mediana`.

- **Sesgo negativo**: muchos valores altos y algunos muy bajos.  
  Resultado: `media < mediana`.

- **Sin sesgo**: los valores est√°n distribuidos de forma sim√©trica.  
  Resultado: `media = mediana`.

---

- `mode()`: devuelve el valor m√°s repetido (moda).
- `median()`: devuelve el valor central (mediana).
- `mean()`: devuelve el promedio (media).

````python
# m√≥dulo para generar n√∫meros aleatorios
import random 

# m√≥dulo para c√°lculos estad√≠sticos
import statistics

# genera una lista de 50 n√∫meros aleatorios del 1-100
# random.randint(1, 100)  --> genera un n√∫mero entero aleatorio
# for i in range(50) --> lo repite 50 veces
numeros_aleatorios = [random.randint(1, 100) for i in range(50)]

# n√∫mero que m√°s se reptire
moda = statistics.mode(numeros_aleatorios)

# n√∫mero que se encuentra en el centro en una lista ordenada
# round lo √∫tilizo para que me redonde√© a 2 decimales
mediana = round(statistics.median(numeros_aleatorios), 2)

# es el promedio - suma de todos los valores dividido por la cantidad
media = round(statistics.mean(numeros_aleatorios), 2)

print("N√∫meros aleatorios:", numeros_aleatorios)
print("Moda:", moda)
print("Mediana:", mediana)
print("Media:", media)

if media > mediana:
    print("Sesgo positivo (asimetr√≠a a la derecha)")
elif media < mediana:
    print("Sesgo negativo (asimetr√≠a a la izquierda)")
else:
    print("No hay sesgo (distribuci√≥n sim√©trica)")

````

### Actividad 7
Escribir un programa que solicite una frase o palabra al usuario. Si termina con vocal, a√±adir un signo de exclamaci√≥n al final e imprimir; en caso contrario, imprimirla tal como fue ingresada.

````python
# Pedir una frase o palabra al usuario
frase = input("Ingres√° una frase o palabra: ")

# Obtener la √∫ltima letra
ultima_letra = frase[-1].lower()

# Verificar si termina en vocal
if ultima_letra in "aeiou":
    # Si termina en vocal, agregar signo de exclamaci√≥n
    frase = frase + "!"
    print("Resultado:", frase)
else:
    # Si no termina en vocal, imprimir tal como est√°
    print("Resultado:", frase)
````

Escribir un programa que pida al usuario la magnitud de un terremoto y lo clasifique seg√∫n la escala de Richter:

- **Mayor o igual a 3 y menor a 4:** Leve  
- **Mayor o igual a 4 y menor a 5:** Moderado  
- **Mayor o igual a 5 y menor a 6:** Fuerte  
- **Mayor o igual a 6 y menor a 7:** Muy fuerte
