### Actividad 10
Escribir un programa que pregunte al usuario el hemisferio (N/S), el mes y el d√≠a. Determinar la estaci√≥n del a√±o seg√∫n la tabla dada.

Estaciones:
- 21/12 a 20/3: Invierno (N), Verano (S)
- 21/3 a 20/6: Primavera (N), Oto√±o (S)
- 21/6 a 20/9: Verano (N), Invierno (S)
- 21/9 a 20/12: Oto√±o (N), Primavera (S)

### üåé Estaciones del a√±o seg√∫n el hemisferio

| Periodo del a√±o                         | Estaci√≥n en el hemisferio norte | Estaci√≥n en el hemisferio sur |
|----------------------------------------|----------------------------------|-------------------------------|
| Desde el 21 de diciembre hasta el 20 de marzo  | Invierno                         | Verano                        |
| Desde el 21 de marzo hasta el 20 de junio     | Primavera                        | Oto√±o                         |
| Desde el 21 de junio hasta el 20 de septiembre| Verano                           | Invierno                      |
| Desde el 21 de septiembre hasta el 20 de diciembre | Oto√±o                        | Primavera                     |


````python

# Solicitar datos al usuario
hemisferio = input("Ingrese el hemisferio (N/S): ").upper()  #se usa la funci√≥n .upper para pasar letras a May√∫scula
mes = int(input("Ingrese el mes (1-12): "))
dia = int(input("Ingrese el d√≠a (1-31): "))

# Determinar la estaci√≥n
if (mes == 12 and dia >= 21) or (mes in [1, 2]) or (mes == 3 and dia <= 20):
    estacion_norte = "Invierno"
    estacion_sur = "Verano"
elif (mes == 3 and dia >= 21) or (mes in [4, 5]) or (mes == 6 and dia <= 20):
    estacion_norte = "Primavera"
    estacion_sur = "Oto√±o"
elif (mes == 6 and dia >= 21) or (mes in [7, 8]) or (mes == 9 and dia <= 20):
    estacion_norte = "Verano"
    estacion_sur = "Invierno"
elif (mes == 9 and dia >= 21) or (mes in [10, 11]) or (mes == 12 and dia <= 20):
    estacion_norte = "Oto√±o"
    estacion_sur = "Primavera"
else:
    estacion_norte = "Desconocida"
    estacion_sur = "Desconocida"

# Mostrar resultado
if hemisferio == "N":
    print(f"Est√°s en el hemisferio norte, y la estaci√≥n es: {estacion_norte}")
elif hemisferio == "S":
    print(f"Est√°s en el hemisferio sur, y la estaci√≥n es: {estacion_sur}")
else:
    print("Hemisferio no v√°lido. Ingrese 'N' o 'S'.")

````
