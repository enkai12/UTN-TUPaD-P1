## 2) Desarrolla un programa que solicite al usuario un número entero y determine la cantidad de dígitos que contiene.

```python
# 2)

# Solicitamos un número al usuario
num = int(input("Ingresá un número entero: "))

# Convertimos el número a texto y quitamos el signo negativo con la funcion abs.
num_cadena = str(abs(num))

# Contamos la cantidad de caracteres que tiene el número con la funcion len.
contador = len(num_cadena)

print("El número tiene", contador, "dígitos.")

```

## 3) Escribe un programa que sume todos los números enteros comprendidos entre dos valores dados por el usuario, excluyendo esos dos valores.

```python
# 3)

# Pedimos dos valores al usuario
inicio = int(input("Ingresá el primer número: "))
fin = int(input("Ingresá el segundo número: "))

# Aseguramos que 'inicio' sea menor que 'fin'
if inicio > fin:
    # Si están invertidos, los intercambiamos
    aux = inicio
    inicio = fin
    fin = aux

suma = 0  # Inicializamos la suma

# Sumamos desde inicio+1 hasta fin-1
for i in range(inicio + 1, fin):
    suma += i  # Acumulamos la suma

print("La suma de los números entre", inicio, "y", fin, "es:", suma)
```

## 4) Elabora un programa que permita al usuario ingresar números enteros y los sume en secuencia. El programa debe detenerse y mostrar el total acumulado cuando el usuario ingrese un 0.

```python
# 4)

suma = 0

# bucle que se repite hasta que se ingrese 0

num = int(input("Ingresá un número entero que se sumaran en secuencia y (0 para terminar el programa): "))
while num != 0:
    suma += num  # Sumamos el número
    num = int(input("Ingresá otro número entero (0 para terminar): "))

print(f"El total acumulado es: , {suma}")
```

## 5) Crea un juego en el que el usuario deba adivinar un número aleatorio entre 0 y 9. Al final, el programa debe mostrar cuántos intentos fueron necesarios para acertar el número.

```python
import random  # librería para números aleatorios

secreto = random.randint(0, 9)  # Generamos el número secreto
intentos = 0  # Contador de intentos

# Usamos un bucle hasta que el usuario adivine
acertado = False
while not acertado:
    intento = int(input("Adiviná el número (entre 0 y 9): "))
    intentos += 1  # Sumamos un intento
    if intento == secreto:
        acertado = True  # Terminamos el bucle
    else:
        print("Incorrecto, intentá de nuevo.")

print(f"Correcto! Adivinaste en {intentos} intentos.")
```

## 7) Crea un programa que calcule la suma de todos los números comprendidos entre 0 y un número entero positivo indicado por el usuario.#

```python
num = int(input("Ingrese un numero entero positivo: ")) # se pide ingresar un numero

if num < 0: # si el num es menor a 0 no es positivo
    print("Debes ingresar un numero positivo")
else:
    suma = 0
    for i in range(0, num + 1): # num + 1 para que tome el rango total del numero ingresado
        suma += i

print(f"La suma de 0 a {num} es: {suma}")
```

## 8) Escribe un programa que permita al usuario ingresar 100 números enteros. Luego, el programa debe indicar cuántos de estos números son pares, cuántos son impares, cuántos son negativos y cuántos son positivos. (Nota: para probar el programa puedes usar una cantidad menor, pero debe estar preparado para procesar 100 números con un solo cambio).

```python
cantidad_num = 5 # se inicializa con la cantidad que crea necesario

pares = 0
impares = 0
negativos = 0
positivos = 0

print(f"Ingresa {cantidad_num} numeros enteros! ")

for i in range(cantidad_num): # se conoce el range y se hace una estructura for para ir recorriendo la cantidad de numeros ingresados
    num = int(input(f"Ingresa un numero {i + 1} "))

    if num % 2 == 0:
        pares += 1
    else:
        impares += 1
    
    if num < 0:
        negativos += 1
    elif num > 0:
        positivos += 1

print(f"\n Resultados: \n Numeros pares: {pares} \n Numeros impares: {impares} \n Positivos: {positivos} \n Negativos: {negativos}")

```

## 9) Elabora un programa que permita al usuario ingresar 100 números enteros y luego calcule la media de esos valores. (Nota: puedes probar el programa con una cantidad menor, pero debe poder procesar 100 números cambiando solo un valor).

```python
cantidad_num = int(input(f"Ingrese cuantos numeros enteros quiere ingresar para calcular la MEDIA: "))

suma = 0

for i in range(cantidad_num):
    num = int(input(f"Ingresa numeros enteros: {i + 1} \n Se calculara la media de los numeros enteros ingresados"))
    suma += 1
    print(f"{num}")

media = suma / cantidad_num

print(f"\nLa media de los {cantidad_num} numeros ingresados es: \nMedia: {media} ")
```

## 10) Escribe un programa que invierta el orden de los dígitos de un número ingresado por el usuario. Ejemplo: si el usuario ingresa 547, el programa debe mostrar 745.

```python
num = input("Ingresa un número: ")

if num.startswith('-'):
    numero_invertido = '-' + num[:0:-1] # invierte la cadena desde la posicion 0 hasta la ultima posicion incluida, el signo '-' no se incluye.
else:
    numero_invertido = num[::-1] # invierte toda la cadena

print(f"\nNúmero ingresado: {num} \nNúmero invertido: {numero_invertido}")
```
